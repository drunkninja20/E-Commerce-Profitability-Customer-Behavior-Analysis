```python
df['Order_Date']=pd.to_datetime(df['Order_Date'])
df['Ship_Date']=df['Ship_Date'].astype('datetime64[ns]')
df.info()
```

```python
df['Order_Date_Year']=df['Order_Date'].dt.year
df['Order_Date_Month']=df['Order_Date'].dt.month
df['Order_Date_Day']=df['Order_Date'].dt.day
```

```python
#annual sales
annual_sales=df.groupby(df['Order_Date_Year'])['Sales'].sum().reset_index()
annual_sales
#set_index not used as annual sales returns a series
plt.plot(annual_sales['Order_Date_Year'],annual_sales['Sales'],marker='o')
for x,y in zip(annual_sales['Order_Date_Year'],annual_sales['Sales']):
    plt.annotate(
        "{:.2f}".format(y),  # Format to 2 decimal places
        (x,y), # Position of the point
        textcoords="offset points",
        xytext=(10, 10),     # Offset the text 10 points above the point
        ha='center',         # Horizontal alignment
        fontsize=9           # Optional: adjust font size
    )
plt.grid()
plt.title('Yearly Sales Trend')

plt.show()
```

```python
#monthly sales 
#dt.to_period = Converts DatetimeArray/Index to PeriodArray/PeriodIndex based on the period aliases such as 'M' which means monthly frequency
df['Year-Month']=df['Order_Date'].dt.to_period('M')
monthly_sales=df.groupby(df['Year-Month'])['Sales'].sum().reset_index()
# ERROR THROWN = TypeError: float() argument must be a string or a real number, not 'Period'
#SOLUTION = Convert year-month to str by using astype
plt.figure(figsize=(14, 5))
plt.plot(monthly_sales['Year-Month'].astype(str),monthly_sales['Sales'],marker='o')

plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
annual_sales=df.groupby('Order_Date_Year')['Sales'].sum().reset_index()
annual_sales['yoy_growth']=annual_sales['Sales'].pct_change()*100
plt.figure(figsize=(10,6))
plt.plot(annual_sales['Order_Date_Year'], annual_sales['yoy_growth'], marker='o')
for x,y in zip(annual_sales['Order_Date_Year'], annual_sales['yoy_growth']):
    plt.annotate(
        "{:.2f}".format(y),  # Format to 2 decimal places
        (x,y),
        textcoords="offset points",
        xytext=(10,10),
        ha='center',
        fontsize=12)
plt.title('Year-over-Year Growth')
plt.xlabel('Year')
plt.ylabel('YoY Growth (%)')
plt.grid(True)
```

```python
#categorical sales trend
sales_by_category=df.groupby(df['Category'])['Sales'].sum().reset_index()
sales_by_category
plt.plot(sales_by_category['Category'],sales_by_category['Sales'],marker='o')
for x,y in zip(sales_by_category['Category'],sales_by_category['Sales']):
    plt.annotate(
        "{:.2f}".format(y),  # Format to 2 decimal places
        (x,y), # Position of the point
        textcoords="offset points",
        xytext=(10, 10),     # Offset the text 10 points above the point
        ha='center',         # Horizontal alignment
        fontsize=9           # Optional: adjust font size
    )
plt.grid()
plt.title('Categoricaly Sales Trend')

plt.show()
```

```python
df['Profit_Margin_%']=(df['Profit']*100/df['Sales'])
```

```python
plt.scatter(df['Sales'], df['Profit'], alpha=0.5, c='green')

# Add reference lines
plt.axhline(0, color='red', linestyle='--')  # Break-even line
plt.axvline(df['Sales'].median(), color='blue', linestyle=':')  # Median sales

# Labels
plt.title('Profit vs Sales (All Products)')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.grid(alpha=0.3)
plt.show()
```

```python
profit_margin_category = df.groupby('Category').agg(
    Total_Sales=pd.NamedAgg(column='Sales', aggfunc='sum'),
    Total_Profit=pd.NamedAgg(column='Profit', aggfunc='sum'),
    Avg_Profit_Margin=pd.NamedAgg(column='Profit_Margin_%', aggfunc='mean')
).reset_index()
profit_margin_category.sort_values('Avg_Profit_Margin',ascending=False)
```

```python
categories = profit_margin_category['Category']
margins = profit_margin_category['Avg_Profit_Margin']
bars = plt.bar(categories, margins)
plt.title('Profit Margin by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Profit Margin (%)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
profit_margin_subcategory=df.groupby(['Category','Sub-Category']).agg(
    Total_Sales=pd.NamedAgg(column='Sales',aggfunc='sum'),
    Total_Profit=pd.NamedAgg(column='Profit',aggfunc='sum'),
    Avg_Profit_Margin=pd.NamedAgg(column='Profit_Margin_%',aggfunc='mean'))
profit_margin_subcategory.sort_values(['Category','Avg_Profit_Margin'],ascending=[True,False])
```

```python

top_subcats = profit_margin_subcategory.nlargest(5, 'Avg_Profit_Margin')
low_subcats = profit_margin_subcategory.nsmallest(5, 'Avg_Profit_Margin')
sns.barplot(data=top_subcats, y='Sub-Category', x='Avg_Profit_Margin')
sns.barplot(data=low_subcats, y='Sub-Category', x='Avg_Profit_Margin')
plt.title('Top 5 Most & Least Profitable Sub-Categories')
plt.xlabel('Average Profit Margin (%)')
plt.ylabel('')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
palette = {'Technology': '#4e79a7', 'Office Supplies': '#f28e2b', 'Furniture': '#e15759'}

bars = sns.barplot(
    data=top_subcats,
    y='Sub-Category',
    x='Avg_Profit_Margin',
    hue='Category',
    dodge=False,
    palette=palette
)

plt.title('Top 5 Most Profitable Sub-Categories (Grouped by Parent Category)', pad=15)
plt.xlabel('Average Profit Margin (%)')
plt.ylabel('')
plt.xlim(0, top_subcats['Avg_Profit_Margin'].max() * 1.2)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
profit_per_product = df.groupby('Product_Name')['Profit'].sum().sort_values(ascending=False)
top_10_products = profit_per_product.head(10).reset_index()

# Horizontal bar plot
sns.barplot(data=top_10_products, y='Product_Name', x='Profit', palette='viridis')
plt.title('Top 10 Most Profitable Products')
plt.xlabel('Total Profit ($)')
```

```python
#day of the week analysis
df['Day_of_Week']=df['Order_Date'].dt.day_name()
daily_orders=df['Day_of_Week'].value_counts().reset_index()
plt.figure(figsize=(8,6))
plt.bar(daily_orders['Day_of_Week'],daily_orders['count'])
for x,y in zip(daily_orders['Day_of_Week'],daily_orders['count']):
    plt.annotate(
        "{:.2f}".format(y),
        (x,y),
        textcoords="offset points",
        xytext=(-15, 8))
plt.title('Orders by the Day of the Week')
plt.show()
```

```python
#profit vs sales for top 10 sales city
#city_sales=df.groupby(['City'])['Sales'].sum().sort_values(ascending=False).reset_index()
#city_profit=df.groupby(['City'])['Profit'].sum().sort_values(ascending=False).reset_index()
#use agg function to calculate sum of sales and profit for each city
top_cities=df.groupby(['City']).agg({'Sales':'sum','Profit':'sum'}).nlargest(10,'Sales').reset_index()
top_city_melt=pd.melt(top_cities,id_vars='City',value_vars=['Sales','Profit'],var_name='Metric',value_name='Amount')
plt.figure(figsize=(13,5))
sns.barplot(data=top_city_melt,x='City',y='Amount',hue='Metric')
plt.title('Top 10 Cities: Sales vs Profit Comparison')
plt.xlabel('City')
plt.ylabel('Amount ($)')
```

```python
df['Cost'] = df['Sales'] - df['Profit']
df['ROI (%)'] = (df['Profit'] / df['Cost']) * 100
roi_by_category = df.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Cost=('Cost', 'sum'),
    Avg_ROI=('ROI (%)', 'mean')
).reset_index()
roi_by_category = roi_by_category.sort_values('Avg_ROI', ascending=False)

plt.figure(figsize=(9, 5))
ax = sns.barplot(
    data=roi_by_category, 
    x='Category', 
    y='Avg_ROI',
    hue='Avg_ROI',
    legend=False
)
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.1f}%", 
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center', 
        xytext=(0, 10), 
        textcoords='offset points'
    )

plt.title("ROI by Product Category")
plt.xlabel("Category")
plt.ylabel("Average ROI (%)")
plt.ylim(0, roi_by_category['Avg_ROI'].max() * 1.3)
plt.tight_layout()
plt.show()
#Passing `palette` without assigning `hue` is deprecated and will be removed in v0.14.0. Assign the `y` variable to `hue` and set `legend=False` for the same effect. 
#sns.barplot(data=roi_subcat, y='Category', x='ROI', palette='viridis')
```

```python

df['ROI'] = (df['Profit'] / (df['Sales'] - df['Profit'])) * 100
roi_subcat = df.groupby('Sub-Category')['ROI'].mean().nlargest(10).reset_index()

plt.figure(figsize=(12, 5))
sns.barplot(data=roi_subcat, y='Sub-Category', x='ROI', palette='viridis',hue='Sub-Category',legend=False)
plt.title('Top 10 Sub-Categories by ROI', fontweight='bold')
plt.xlabel('ROI (%)')
for i, v in enumerate(roi_subcat['ROI']):
    plt.text(v+1, i, f"{v:.1f}%", va='center')
plt.show()
```

```python
#clv
clv = df.groupby('Customer_ID').agg(
    first_purchase=('Order_Date', 'min'),
    last_purchase=('Order_Date', 'max'),
    frequency=('Order_ID', 'count'),
    avg_purchase=('Sales', 'mean')
)
clv['lifespan'] = (clv['last_purchase'] - clv['first_purchase']).dt.days / 365
clv['CLV'] = clv['avg_purchase'] * clv['frequency'] * clv['lifespan']
clv.nlargest(10, 'CLV')['CLV'].plot(kind='barh', figsize=(10, 5))
plt.title('Top 10 Customers by CLV')
plt.xlabel('Customer Lifetime Value ($)')
plt.show()
```

```python
#Pareto Chart
product_value = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
product_value_cumsum = product_value.cumsum() / product_value.sum()

fig, ax = plt.subplots(figsize=(12, 5))
product_value.plot(kind='bar', ax=ax)
ax2 = ax.twinx()
product_value_cumsum.plot(ax=ax2, marker='o')
ax.axhline(y=product_value.sum()*0.8, linestyle='--')
ax.set_title('Pareto Analysis: Product Sales Contribution')
ax.set_ylabel('Total Sales ($)')
ax2.set_ylabel('Cumulative Percentage')
plt.show()
```

```python
# time series forecasting
monthly_sales = df.set_index('Order_Date')['Sales'].resample('M').sum()
plt.figure(figsize=(12, 5))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.ylabel('Sales ($)')
plt.grid()
z = np.polyfit(range(len(monthly_sales)), monthly_sales.values, 1)
p = np.poly1d(z)
plt.plot(monthly_sales.index, p(range(len(monthly_sales))), 'o')
plt.show()
```
